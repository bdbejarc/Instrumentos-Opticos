```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Funci√≥n para graficar intensidad, amplitud y fase
def ploteo_complejo(mat, indicador, escala):
  if(indicador == "I"):
    mat = np.abs(mat)**2
  elif(indicador == "A"):
    mat = abs(mat)
  elif (indicador == "P"):
    mat = np.angle(mat)

  if (escala == 1):
    plt.figure()
    plt.imshow(mat)
    plt.colorbar()
    plt.show()

  else:
    plt.figure()
    plt.imshow(np.log(mat+0.0000001))
    plt.colorbar()
    plt.show()

  return 0
```

```python
#Carga y transformada de Fourier de la imagen en estudio
img = cv2.imread('b.png',0)
fft_img_prev = np.fft.fftn(img)
fft_img = np.fft.fftshift(fft_img_prev)

plt.figure()
plt.imshow(img)
plt.colorbar()
plt.show()

np.shape(img)
```

```python
#Gr√°fica de la transformada de Fourier de la imagen 
ploteo_complejo(fft_img, "I", 0)
```

```python
#Descargar imagen
#Normalizando
i = (np.abs(fft_img))**2
i_norm = 255 * (i - np.min(i))/(np.max(i) - np.min(i))
cv2.imwrite("FFT.jpg", np.uint8(np.log(i_norm+0.00001)))
```

```python
#M√°scara circular
mask = np.zeros(np.shape(img))
coord = int(np.shape(img)[0]/2)
print (coord)
cv2.circle(mask, (coord, coord), 25, 10, -1)
plt.imshow(mask)
plt.colorbar()
plt.show()
```

```python
cv2.imwrite("cir.jpg", np.uint8(np.abs(mask))**2)
```

```python
new_img = np.fft.fftn(mask * fft_img);
```

```python
ploteo_complejo(new_img, "I", 1)
```

```python
#Normalizando
intensidad = (np.abs(new_img))**2
intensidad_norm = 255 * (intensidad - np.min(intensidad))/(np.max(intensidad) - np.min(intensidad))
cv2.imwrite("cir2.jpg", np.uint8(intensidad_norm))
```
