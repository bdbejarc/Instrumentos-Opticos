```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def ploteo_complejo(mat, indicador, escala):
  if(indicador == "I"):
    mat = np.abs(mat)**2
  elif(indicador == "A"):
    mat = abs(mat)
  elif (indicador == "P"):
    mat = np.angle(mat)

  if (escala == 1):
    plt.figure()
    plt.imshow(mat)
    plt.colorbar()
    plt.show()

  else:
    plt.figure()
    plt.imshow(np.log(mat+0.0000001))
    plt.colorbar()
    plt.show()

  return 0
```

```python
img = cv2.imread('b.png',0)
fft_img_prev = np.fft.fftn(img)
fft_img = np.fft.fftshift(fft_img_prev)

plt.figure()
plt.imshow(img)
plt.colorbar()
plt.show()

np.shape(img)
```

```python
ploteo_complejo(fft_img, "I", 0)
```

```python
cv2.imwrite("in.jpg", np.uint8(np.abs(fft_img))**2)
```

```python
mask = np.ones(np.shape(img))
coord = int(np.shape(img)[0]/2)
print (coord)
cv2.circle(mask, (coord, coord), 50, 0, -1)
plt.imshow(mask)
plt.colorbar()
plt.show()
```

```python
new_img = np.fft.fftn(mask * fft_img);
```

```python
ploteo_complejo(new_img, "I", 1)
```
